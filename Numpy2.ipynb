# np.linspace :
      used to create a sequence of numbers with constant spacing or linear spacing between 2 numbers.

```python
np.linspace(1,15)
```

```python
np.linspace(1,15,num=20)
```

```python
np.linspace(1,15,num=20,endpoint =False)
```

```python
 np.linspace(1,15,retstep=True,endpoint=False)   #retstep returns difference between two values
```

# Sin/cos/tan wave using linspace

```python
inp = np.linspace(2,22,num= 100)
out = np.sin(inp)
out
```

```python
import matplotlib.pyplot as plt #data visualization library
plt.plot(inp,out,'g--')
```

```python
#tan plot
inp1 = np.linspace(2,22,num=50)
out2 = np.tan(inp1)
plt.plot(inp1,out2,'r--')
```

```python
inp3=np.linspace(1,20,num=30)
out3=np.cos(inp3)
plt.plot(inp3,out3,'b--')
```

# where(): condition based conversion we can carry out

```python
p= np.arange(24).reshape(6,4)
```

```python
np.where(p>20,1,p)
```
